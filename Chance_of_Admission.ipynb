```python
#Import Data
# Reading the data set using pandas

df = pd.read_csv('/content/Student Performance Predictor for EduQuest Coaching.csv')

# Here I am using head to display the first 5 rows.

df.head()
```

```python
#Data Visualization
# Info is used here so as to display info like data type, number of null values and total rows.
df.info()

# Drops the rows with null values.
df = df.dropna()

# Describe Data
df.describe(include='all')
```

```python
#Data Preprocessing
# Replacing Females with Number 1 and Male with no. 0

df['gender'] = df['gender'].replace({'Female': 1, 'Male': 0})

# Parents education level is replaced by numbers. Diploma with 0, high school with 1, graduate with 2 and post grad with 3.
df['parental_education'] = df['parental_education'].replace({
    'Diploma': 0,
    'High School': 1,
    'Graduate': 2,
    'Postgraduate': 3
})

# Checking Internet Access

df['internet_access'] = df['internet_access'].replace({'Yes': 1, 'No': 0})

# Replacing tutor support (yes=1 & no=0)

df['tutor_support'] = df['tutor_support'].replace({'Yes': 1, 'No': 0})

# Extra Curricular Involvement is also replaced with low, med, high.

df['extra_curricular_involvement'] = df['extra_curricular_involvement'].replace({
    'Low': 0,
    'Moderate': 1,
    'High': 2
})
```

```python
#Define Target Variable (y) and Feature Variables (X)
# y is the variable that has to be predicted.
y = df['final_exam_score']

# x is the other set of variables that are used to predict y.

X = df[['gender', 'age', 'parental_education', 'family_income',
        'internet_access', 'previous_exam_score', 'attendance_rate',
        'homework_completion_rate', 'class_participation_score',
        'number_of_absences', 'extra_curricular_involvement',
        'learning_hours_per_week', 'tutor_support']]
```

```python
#Train Test Split
#training set is used train the model
#testing set is used to predict the unseen data.

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2529)
```

```python
#Modeling
# linear regression is implemented
model = LinearRegression()

#Model Evaluation
# training the model
model.fit(X_train, y_train)
```

```python
# x_test is used to calculate y_pred.
y_pred = model.predict(X_test)
```

```python
#Prediction
# Output of Mean Absolute Percentage Error is obtained.
mean_absolute_percentage_error(y_test, y_pred)
```
